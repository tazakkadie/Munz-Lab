## set the parameters

```python
# general dataset-dependent parameters
fr = 30                     # imaging rate in frames per second
decay_time = 0.4            # length of a typical transient in seconds
dxy = (2., 2.)              # spatial resolution in x and y in (um per pixel)

# motion correction parameters
strides = (48, 48)          # start a new patch for pw-rigid motion correction every x pixels
overlaps = (24, 24)         # overlap between patches (width of patch = strides+overlaps)
max_shifts = (6,6)          # maximum allowed rigid shifts (in pixels)
max_deviation_rigid = 3     # maximum shifts deviation allowed for patch with respect to rigid shifts
pw_rigid = True             # flag for performing non-rigid motion correction

# CNMF parameters for source extraction and deconvolution
p = 1                       # order of the autoregressive system (set p=2 if there is visible rise time in data)
gnb = 2                     # number of global background components (set to 1 or 2)
merge_thr = 0.85            # merging threshold, max correlation allowed
bas_nonneg = True           # enforce nonnegativity constraint on calcium traces (technically on baseline)
rf = 15                     # half-size of the patches in pixels (patch width is rf*2 + 1)
stride_cnmf = 10             # amount of overlap between the patches in pixels (overlap is stride_cnmf+1) 
K = 4                       # number of components per patch
gSig = np.array([4, 4])     # expected half-width of neurons in pixels (Gaussian kernel standard deviation)
gSiz = 2*gSig + 1           # Gaussian kernel width and hight
method_init = 'greedy_roi'  # initialization method (if analyzing dendritic data see demo_dendritic.ipynb)
ssub = 1                    # spatial subsampling during initialization 
tsub = 1                    # temporal subsampling during intialization

# parameters for component evaluation
min_SNR = 2.0               # signal to noise ratio for accepting a component
rval_thr = 0.85             # space correlation threshold for accepting a component
cnn_thr = 0.99              # threshold for CNN based classifier
cnn_lowest = 0.1            # neurons with cnn probability lower than this value are rejected
```

## extract where your time and data colums are

```python
# Extract time column and data columns
time = data.iloc[:, 0].values            # time: first column
data = data.iloc[:, 1:7].values          # data: next 6 ROI columns

# Use ROI1, ROI2, ROI3, etc. as column names
columns = [f"ROI{i}" for i in range(1, data.shape[1] + 1)]

# Create the TsdFrame
tsdframe= nap.TsdFrame(t=time, d=data, columns=columns)

print(tsdframe)
```

## Apply bandpass filter
- Change the lower and upper cutoffs according to your needs. Max upper cutoff is fs/2
- Set your imaging/signal frequency (fs = ?)

```python
import pynapple as nap

# filtering certain frequencies
clean_tsdframe = nap.apply_bandpass_filter(tsdframe, (0.01, 0.45), fs=1, mode='butter')

print(clean_tsdframe)
```

## Bandpass filter effect
Set to 'python' environment!
1. Raw calcium traces
2. Calcium traces after the bandpass filter is applied

```python
# plotting to compare before vs after bandpass filter

import matplotlib.pyplot as plt

fig, axs = plt.subplots(2,1)

plt.sca(axs[0])
plt.plot(tsdframe, linewidth = 1.5, label = "before filter")

plt.sca(axs[1])
plt.plot(clean_tsdframe, linewidth = 1.5, label = 'after filter')

plt.tight_layout()
plt.show()
```

## Compute the Fast Fourier Transform (FFT)
Set your imaging frequency (dt = ?)

```python
# compute the fast fourier transform (FFT)

dt = 0.672                                                  # no. of frames / duration of recording (frame rate)

n = len(time)
fhat = np.fft.fft(clean_tsdframe.to_numpy(), n)             # compute the FFT
PSD = fhat * np.conj(fhat) / n                              # power spectrum (power per frequency)
freq = (1/(dt * n)) * np.arange(n)                          # create x-axis of frequencies
L = np.arange(1, np.floor(n/2), dtype = 'int')              # only plot the first half
```

## Power Density Spectrum (PSD)
- y-axis = power/strength of signal the frequency 
- x-axis = frequencies

When there is a large spike at some value on the x-axis, that means this frequency is prominent in the traces.

```python
plt.plot(freq[L], PSD[L], color = 'b', linewidth = 2)
plt.show()
```

## Filter out the noise
- Set your threshold as to which frequencies you want to keep.
- Reduce the amplitude of all frequencies within the trace by 'scaling'. Set the scaling power (default at 0.1)

```python
# use the PSD to filter out noise

indices = PSD > 0.02                      # find all freqs with large power
PSDclean = PSD * indices                    # find out all others
fhat = indices * fhat                       # zero out small fourier coeff. in Y
fhat_scaled = 0.1 * fhat
ffilt = np.fft.ifft(fhat_scaled)                   # inverse FFT for filtered time signal

plt.plot(time, ffilt, color = 'k', linewidth = 2, label = 'filtered')
plt.show()
```

## Visualize the calcium traces after processing
1. Calcium traces after bandpass filter
2. Power Spectrum showing which frequencies we have left (in red).
3. Calcium traces with only the desired frequencies left.

```python
# plots

fig, axs = plt.subplots(3,1)
plt.style.use('seaborn-v0_8-notebook')

plt.sca(axs[0])
plt.title('Calcium Traces')
plt.plot(clean_tsdframe, linewidth = 1.5, label = 'noisy')


plt.sca(axs[1])
plt.title('Power Spectrum')
plt.plot(freq[L], PSD[L], linewidth = 2)
plt.plot(freq[L], PSDclean[L], linewidth = 1.5, label = 'filtered', color = 'red')
plt.xlim(freq[L[0]], freq[L[-1]])


plt.sca(axs[2])
plt.title('all the main frequencies filtered out')
plt.plot(time, ffilt, linewidth = 2, label = 'filtered', color="blue")
plt.legend

plt.tight_layout()
plt.show()
```

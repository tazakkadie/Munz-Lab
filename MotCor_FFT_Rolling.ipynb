```python
#load data with numpy 
#since we're measuring intensities via ROIs and then doing FFT, we only need to use a 1D FFT (if we were doing it on an image, we'd need a 2D FFT)
import numpy as np 
import matplotlib.pyplot as plt

# if encoutering \ufeff when using np.loadtxt, that means it's encountering a Byte Order Mark (BOM) at the beginning of your file so use np.genfromtxt and input encoding='utf-8-sig' as an argument 
#usecols(np.arange()) if you want to specify a range of columns to load  
signal = np.genfromtxt('/Users/Jeremy/Desktop/test.csv', encoding='utf-8-sig', delimiter = ',', usecols=np.arange(0,1), dtype='f8')
#signal = pd.read_csv('/Users/Jeremy/Desktop/test.csv',usecols=[1], header=None ) #use square brackets to call 1 column, round brackets to call more than 1 column. NOTE pandas dataframe does not work for FFT 
#print(signal)

dt = 0.522 #samples per second 
t = np.arange(0,179.6,dt) #creates a range that allows floats (for precision), where 0 is the start, 179.6 is the stop (which isn't included in the range), and dt is the step size 

plt.plot(t,signal)
plt.title('Raw')
plt.xlabel("Time (s)")
plt.ylabel("Intensity")

```

```python
#visualizing FFT 
n=len(t)                            #len() counts the number of items in t
fhat = np.fft.fft(signal,n)

plt.subplot(1,2,1)
plt.plot(np.fft.fftshift(np.real(fhat)))
plt.title("real FFT")
plt.ylabel("Magnitude")
plt.xlabel('frequency index')
#plt.xlim(100, 250)

plt.subplot(1,2,2)
plt.plot(np.fft.fftshift(np.imag(fhat)))
plt.title("imaginary FFT")
plt.ylabel("Magnitude")
plt.xlabel('frequency index')

plt.subplots_adjust(wspace=0.3)

#if it looks funny, it may be because it's computing multiple ROIs simultaneously
```

```python
#FFT with numpy 
n=len(t) #len() counts the number of items in t
print(n)

fhat = np.fft.fft(signal,n)     #compute the FFT
PSD = fhat * np.conj(fhat)/n    #Power spectrum density (gives you the magnitude of each fourier coefficient and we'll get a vector of powers at each frequency)
freq = (1/(dt*n))*np.arange(n)    #create an x-axis of frequencies 
L = np.arange(1,np.floor(n/2),dtype='int') #only plot first half (real signals are symmetric)


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8,8))
x1=t
x2=freq[L]
y1=signal
y2=PSD[L]

ax1.plot(x1, y1)
ax1.set_title("Raw")
ax1.set_xlabel('Time(s)')
ax1.set_ylabel('Fluorescence')

ax2.plot(x2, y2)
ax2.set_title("PSD")
ax2.set_xlabel("Freq (Hz)")
ax2.set_ylabel("Amplitude")

plt.subplots_adjust(hspace=0.3)
```

```python
#determine mean amplitude of PSD(?)
amplitudes = np.sqrt(PSD) #is this right?? 
a1 = np.real(PSD)
a1_s = np.sqrt(a1)

mean_amplitude = np.mean(amplitudes) 
mean1 = np.mean(a1)
mean1s = np.mean(a1_s)
print(mean_amplitude)
print(mean1)
print(mean1s)
```

```python
for i in range(len(PSD[L])):
    current_value = PSD[i]

    if current_value > mean1s*1.5:
        PSD[i] = current_value * 0  
    elif current_value < mean1s/2:
        PSD[i] = current_value * 0 

filt_sig = np.fft.ifft(PSD)
```

Using deltaF/F to normalize data over baseline 

```python
#Use Scipy to create a rolling baseline 
import numpy as np
from scipy.ndimage import generic_filter 

#parameters
window = 38 # Moving window size (20 s for 1.9 frames/s)
percentile_value = 20 # 20th percentile

#apply moving percentage filter 
F0 = generic_filter(rolling_mean, function=lambda x: np.percentile(x, percentile_value), size=window)

#noramlize data with F0 
dF_over_F = (rolling_mean - F0) / F0 #somehow figure out how to subtract all values from a single column with by F0 

fig, ax = plt.subplots(1,1, figsize=(8,8))                                    
x = time
y=dF_over_F
ax.plot(x,y)
ax.set_title('DeltaF/F')
ax.set_xlabel('Time(s)')
ax.set_ylabel('Intensity')
```

```python
#Use Numpy to create percentage-based baseline 
import numpy as np 
import matplotlib.pyplot as plt 

F0 = np.nanmedian(rolling_mean) #can use median because there's not a lot of ongoing activity (since there's not a lot of spikes that would pull the median above true baseline), otherwise would need to identify a specific percentile 
print(F0)

dF_over_F = (rolling_mean - F0) / F0*100 #turned into percentage 

plt.plot(time, dF_over_F)
plt.xlabel("Time (s)")
plt.ylabel("% DF/F")
plt.title("Signal after FFT, Rolling, and DF/F")
plt.ylim(0, 100)
```

Define what a calcium event is: fluctuations of deltaF/F above baseline with a length greater than 1.5s and a minimum average activity of at least 10%(deltaF/F)/s
Need to calculate the frequency of events too? 

```python
import numpy as np

thresholded = dF_over_F

for i in range(len(thresholded)):
    value = thresholded[i]

    if value < 15:
        thresholded[i] = value * 0  

plt.plot(time, thresholded)
plt.title("True activity")
plt.ylabel(f"% dF/F")
plt.xlabel("Time(s)")
```

Computing integration (AUC) for (%deltaF/F)/s? 

```python
from scipy.integrate import simpson

t = np.arange(0, endpoint, dt)
auc = simpson(thresholded, t)
print(auc)

plt.plot(t, thresholded, label='%F/F')
plt.fill_between(t, thresholded, alpha=0.3, label='Area under curve') #alpha used for blending (values from 0-1)

plt.xlabel('Time (s)')
plt.ylabel('%F/F')
plt.title('Calcium Trace with Area Under Curve')
plt.legend()
plt.tight_layout()
plt.show()
```

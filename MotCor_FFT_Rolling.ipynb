Denoising: After ROI selection and measurement in Fiji, perform FFT (Fast Fourier Transform), then deltaF/F, then rolling mean 

```python
#load data with numpy 
#since we're measuring intensities via ROIs and then doing FFT, we only need to use a 1D FFT (if we were doing it on an image, we'd need a 2D FFT)
import numpy as np 
import matplotlib.pyplot as plt

# if encoutering \ufeff when using np.loadtxt, that means it's encountering a Byte Order Mark (BOM) at the beginning of your file so use np.genfromtxt and input encoding='utf-8-sig' as an argument 
#usecols(np.arange()) if you want to specify a range of columns to load  
signal = np.genfromtxt('/Users/Jeremy/Desktop/test.csv', encoding='utf-8-sig', delimiter = ',', usecols=np.arange(1,2))
#signal = pd.read_csv('/Users/Jeremy/Desktop/test.csv',usecols=[1], header=None ) #use square brackets to call 1 column, round brackets to call more than 1 column. NOTE pandas dataframe might not work for this 
print(signal)

dt = 0.522 #samples per second 
t = np.arange(0,179.6,dt) #creates a range that allows floats (for precision), where 0 is the start, 179.6 is the stop (which isn't included in the range), and dt is the step size 

plt.plot(t,signal)
plt.title('Raw')
plt.xlabel("Time (s)")
plt.ylabel("Intensity")

```

```python
#visualizing FFT 
n=len(t)                            #len() counts the number of items in t
fhat = np.fft.fft(signal,n)

plt.subplot(1,2,1)
plt.plot(np.fft.fftshift(np.real(fhat)))
plt.title("real FFT")
plt.ylabel("Magnitude")
plt.xlabel('frequency index')
#plt.xlim(100, 250)

plt.subplot(1,2,2)
plt.plot(np.fft.fftshift(np.imag(fhat)))
plt.title("imaginary FFT")
plt.ylabel("Magnitude")
plt.xlabel('frequency index')

plt.subplots_adjust(wspace=0.3)

#if it looks funny, it may be because it's computing multiple ROIs simultaneously
```

```python
#FFT with numpy 
n=len(t) #len() counts the number of items in t
print(n)

fhat = np.fft.fft(signal,n)     #compute the FFT
PSD = fhat * np.conj(fhat)/n    #Power spectrum density (gives you the magnitude of each fourier coefficient and we'll get a vector of powers at each frequency)
freq = (1/(dt*n))*np.arange(n)    #create an x-axis of frequencies 
L = np.arange(1,np.floor(n/2),dtype='int') #only plot first half (real signals are symmetric)


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8,8))
x1=t
x2=freq[L]
y1=signal
y2=PSD[L]

ax1.plot(x1, y1)
ax1.set_title("Raw")
ax1.set_xlabel('Time(s)')
ax1.set_ylabel('Fluorescence')

ax2.plot(x2, y2)
ax2.set_title("PSD")
ax2.set_xlabel("Freq (Hz)")
ax2.set_ylabel("Amplitude")

plt.subplots_adjust(hspace=0.3)
```

```python
#determine mean amplitude of PSD(?)
amplitudes = np.sqrt(PSD) #is this right?? 
a1 = np.real(PSD)
a1_s = np.sqrt(a1)

mean_amplitude = np.mean(amplitudes) 
mean1 = np.mean(a1)
mean1s = np.mean(a1_s)
print(mean_amplitude)
print(mean1)
print(mean1s)
```

```python
for i in range(len(PSD[L])):
    current_value = PSD[i]

    if current_value > mean1s*1.5:
        PSD[i] = current_value * 0  
    elif current_value < mean1s/2:
        PSD[i] = current_value * 0 

filt_sig = np.fft.ifft(PSD)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8,8))
plt.subplots_adjust(hspace=0.5)
x1=t
x2=freq[L]
y1=signal
y2=PSD[L]
y3=filt_sig

ax1.plot(x1, y1)
ax1.set_title("Raw")
ax1.set_xlabel('Time(s)')
ax1.set_ylabel('Fluorescence')

ax2.plot(x2, y2)
ax2.set_title("PSD_filtered")
ax2.set_xlabel("Freq (Hz)")
ax2.set_ylabel("Amplitude")

ax3.plot(x1, y3)
ax3.set_title("Filtered_Signal")
ax3.set_xlabel("Time (s)")
ax3.set_ylabel("Fluorescence")

```

```python
#Determine threshold value (can be a right above the noise floor or using a quartile approach)
#Try to set the threshold at 150% of the mean amplitude component of the trace 
#Set all FFT coefficients below the threshold to 0 and perform a reverse FFT to reconstruct signal without excluded frequencies

indices = PSD > mean1s*1.5
PSD_clean = PSD * indices
fhat_filt = indices * fhat 
ffilt = np.fft.ifft(fhat_filt)

complex_ifft_output = np.array(ffilt)
float_array = np.real(complex_ifft_output)
print(float_array)
#np.savetxt('ffilt.csv', float_array, delimiter=',') #to save filtered signal 

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8,8))
x1=freq[L]
x2=t
y1=PSD[L]
y2=PSD_clean[L]
y3=ffilt        #can input float_array as well but output seems the same so I'm not sure what the difference is 

ax1.plot(x1, y1)
ax1.set_title("PSD")
ax1.set_xlabel('Freq(Hz)')

ax2.plot(x1, y2)
ax2.set_title("PSD_clean")
ax2.set_xlabel("Freq (Hz)")
ax2.set_ylabel("Magnitude")

ax3.plot(x2, y3)
ax3.set_title('new_signal')
ax3.set_xlabel('Time(s)')
ax3.set_ylabel('Intensity')


plt.subplots_adjust(hspace=0.5)
```

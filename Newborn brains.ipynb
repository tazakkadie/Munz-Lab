```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Helvetica'

fig, ax = plt.subplots()

# Data per treatment/control group in ____
categories = ['Piriform Ctx', 'CPu Border', 'Hypothalamus', 'Neocortex']
values1 = [133,122,246,0]
values2 = [23,0,29,7]

x = np.arange(len(categories))
ax.bar(x + 0.18, values1,
       width=0.35,
       color='#a91d30',
       alpha=0.9,
       label='cFos+')

ax.bar(x - 0.18, values2,
       width=0.35,
       color='#8ebf4a',
       alpha=0.9,
       label='Arc+')


# Hide top and right spines only
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Labels & title
plt.ylabel('# cells')
plt.xticks(x, categories)
plt.yticks([0,50,150,250]) 
plt.title('During C-Section',
          fontweight='bold')

plt.tight_layout()
plt.legend(frameon=False,
           fontsize=12)

plt.savefig('/Users/tazakkaibrahimadie/Downloads/During C-Section chart.png', 
            dpi=300,
            transparent=True)
plt.show()
```
